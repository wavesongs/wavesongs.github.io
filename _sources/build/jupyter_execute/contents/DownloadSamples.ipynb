### 游댍 Query Audios

First, create a query for the desired sounds. To do this, you need the scientific name of the specie. You can also include additional details such as the country or location where the recording was made to refine your search. Check the [Xeno Canto API](https://xeno-canto.org/explore/api) to check other possible filters.

You can also query multiple species by providing a list of common names along with their corresponding scientific names.

```python
specie_names = [
  ["Rufous-collared Sparrow", "Eurasian Three-toed"],
  ["Zonotrichia capensis", "Picoides tridactylus"]
]
```

````python
from wavesongs.data.xc import download_audios, query_audios

# Define the query parameters
info = {
  "cnt": "Colombia",
  "loc": "Bogota",
}

df_query = query_audios(
    specie_names = ["Rufous-collared Sparrow", "Zonotrichia capensis"],
    info = info,
    max_nb_files=None, # Number of files to download
    format_time=True,
    format_date=True  
)
````

````python
df_query
````

### 拘勇 Download Audios

Define the download directory (`rootdir`) for the audio files selected in the query using the dataset parameters. If no dataset name is specified, the default name will be the scientific name, with spaces replaced by underscores (_).

````python
# Download the first 5 files
no_files = 5
df_query_short = df_query.iloc[:no_files]

df_audios = download_audios(
    df_query_short,
    rootdir = "./assets/audio", 
    dataset_name = '',
    overwrite = True,
    save_csv = True,
    verbose = True
)
````
